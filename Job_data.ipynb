```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

job_data = [] 

for i in range(1, 100):
    url = f'https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=python&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&pDate=I&sequence={i}&startPage={i}'
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    
    jobs = soup.find_all('li', class_="clearfix job-bx wht-shd-bx")
    for job in jobs:
        try:
            job_role = job.find('h2', class_="heading-trun").text.strip()
            company_name = job.find('h3', class_="joblist-comp-name").text.strip()
            updated = job.find('span', class_="sim-posted").text.strip()
            job_description = job.find('li', class_="job-description__").text.strip()
            skill_require = job.find('div', class_="more-skills-sections").text.strip()
            skill_require = ', '.join(word for word in skill_require.split() if word)
            job_location = job.find('li', class_="srp-zindex location-tru").text.strip()

            job_data.append({
                'Company name': company_name,
                'Job Description': job_description,
                'Job Role': job_role,
                'Skills': skill_require,
                'Location': job_location,
                'Updated': updated     
            })
        except AttributeError:
            
            continue

dataset = pd.DataFrame(job_data)


```
